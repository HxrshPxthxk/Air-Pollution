```python
city_data = pd.read_csv("E:\Downloads\stations.csv", index_col = False)
city_data.rename(columns={'Unnamed: 0':'Date'}, inplace = True)
city_data.set_index('Date', inplace=True)
city_data.head()
```

```python
AP = AP001[["PM2.5","PM10","NO2","SO2","O3"]]
newAp = AP.to_numpy()
newAp
```

```python
newAp.shape
```

```python
from numpy import array
def split_sequences(sequences, n_steps):
	X, y = list(), list()
	for i in range(len(sequences)):
		end_ix = i + n_steps
		if end_ix > len(sequences)-1:
			break
		seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix, :]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
n_steps = 5
X, y = split_sequences(newAp, n_steps)
print(X.shape, y.shape)
```

```python
from numpy import hstack
from keras.models import Model
from keras.models import Sequential
from keras.layers import Input
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
```

```python
n_features = 5
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(n_steps, n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(n_features))
model.compile(optimizer='adam', loss='mse')
```

```python
n_features = X.shape[2]
y1 = y[:, 0].reshape((y.shape[0], 1))
y2 = y[:, 1].reshape((y.shape[0], 1))
y3 = y[:, 2].reshape((y.shape[0], 1))
y4 = y[:, 3].reshape((y.shape[0], 1))
y5 = y[:, 4].reshape((y.shape[0], 1))
visible = Input(shape=(n_steps, n_features))
cnn = Conv1D(filters=64, kernel_size=2, activation='relu')(visible)
cnn = MaxPooling1D(pool_size=2)(cnn)
cnn = Flatten()(cnn)
cnn = Dense(50, activation='relu')(cnn)
output1 = Dense(1)(cnn)
output2 = Dense(1)(cnn)
output3 = Dense(1)(cnn)
output4 = Dense(1)(cnn)
output5 = Dense(1)(cnn)
model = Model(inputs=visible, outputs=[output1, output2, output3, output4, output5])
model.compile(optimizer='adam', loss='mse')
model.fit(X, [y1,y2,y3,y4,y5], epochs=200,verbose=0)
```

```python
def result_list(v,n,r):
  start = v
  pre_true = []
  actual_true = []
  PM = [[],[]]
  Pm10 = [[],[]]
  No2 = [[],[]]
  So2 = [[],[]]
  O3 = [[],[]]
  for i in range(0,r):
    x_input = (AP[start:(start+n)].to_numpy())
    x_input = x_input.reshape((1, n_steps, n_features)) 
    yhat = model.predict(x_input, verbose=0)
    from math import sqrt,fabs
    actual_d = list(list(AP[(start+n):(start+n+1)].to_numpy())[0])
    predicted_nr = [list(yhat[i][0])[0] for i in range(0,5)]  
    predicted_d = [round(fabs(num),2) for num in predicted_nr]
    PM[0].append(actual_d[0])
    PM[1].append(predicted_d[0])
    Pm10[0].append(actual_d[1])
    Pm10[1].append(predicted_d[1])
    No2[0].append(actual_d[2])
    No2[1].append(predicted_d[2])
    So2[0].append(actual_d[3])
    So2[1].append(predicted_d[3])
    O3[0].append(actual_d[4])
    O3[1].append(predicted_d[4])
    actual_true.append(list(actual_d))
    pre_true.append(list(predicted_d))
    start += 1
  return actual_true , pre_true , PM , Pm10 , No2 , So2 , O3

actual , predicted , PM , PMT , NO2 , SO2 , O3 = result_list(496,5,100)
print(actual)
print(predicted)
print(PM)
```

```python
l = ["PM","PMT","NO2","SO2","O3"]
ll = [PM,PMT,NO2,SO2,O3]
def plotting_all_data(l):
  for i in range(0,5):
    j = ll[i][0]
    k = ll[i][1]
    plt.plot(j, color ='tab:blue',label='Actual '+l[i]) 
    plt.plot(k, color ='tab:orange', label="Predicted "+l[i])
    plt.title('Comparsion of Actual versus Predicted data in CNN') 
    plt.legend(loc="upper right")
    plt.show()
plotting_all_data(l)
```

#Mean Absolute Error of the Data

```python
def error_in_mae_rmse(ll,l):
  mae = []
  rmse = []
  index = []
  for i in range(0,5):
    mae.append(round(skm.mean_absolute_error(ll[i][0],ll[i][1]),2))
    rmse.append(round(sqrt(skm.mean_squared_error(ll[i][0],ll[i][1])),2))
    index.append(l[i])
  return mae,rmse , index
mae , rmse , index = error_in_mae_rmse(ll,l)
print(mae,rmse,index)
```

```python
data = {'MAE':mae,'RMSE':rmse}
mae_rmse_df = pd.DataFrame(data,index=index)
mae_rmse_df
```
