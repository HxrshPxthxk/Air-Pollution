```python
def unbox(a):
  l = []
  for i in range(len(a)):
    l.append(a[i][0])
  return l
```

```python
k = ["PM2.5","PM10","NO2","SO2","O3"]
```

```python
from numpy import array
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```python
def all_LSTM(n_features,n_steps,index_k):
  newAp1 = AP001[[k[index_k]]]
  newAp1 = array(newAp1)
  newAp1 = unbox(newAp1)
  train_data = newAp1[:500]
  test_data = newAp1[500:605]
  mn = min(test_data)
  mx = max(test_data)
  X , Y = split_sequence(train_data,n_steps)
  model = Sequential()
  model.add(LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
  model.add(Dense(1))
  model.compile(optimizer='adam', loss='mse')
  X = X.reshape((X.shape[0], X.shape[1], n_features))
  model.fit(X, Y, epochs=200, verbose=0)
  x2 , y2 = split_sequence(test_data,n_steps)
  y_pred = []
  for x_input in x2:
    x_input = x_input.reshape((1, n_steps, n_features))
    yhat = model.predict(x_input, verbose=0)
    y_pred.append(yhat)

  return unbox(unbox(y_pred)) , y2 , mn , mx
```

```python
PM_pred , PM_y , m1 , m2= all_LSTM(1,5,0)
PM10_pred , PM10_y , m3 ,   m4= all_LSTM(1,5,1)
NO2_pred , NO2_y , m5 ,m6= all_LSTM(1,5,2)
SO2_pred , SO2_y , m7, m8 = all_LSTM(1,5,3)
O3_pred , O3_y ,m9 , m10 = all_LSTM(1,5,4)
```

```python
from math import sqrt
from sklearn.metrics import mean_absolute_error as Mae
from sklearn.metrics import mean_squared_error as Mse
mae = [Mae(PM_pred,PM_y),Mae(PM10_pred,PM10_y),Mae(NO2_pred,NO2_y),Mae(SO2_pred,SO2_y),Mae(O3_pred,O3_y)]
rmse = [sqrt(Mse(PM_pred,PM_y)),sqrt(Mae(PM10_pred,PM10_y)),sqrt(Mae(NO2_pred,NO2_y)),sqrt(Mae(SO2_pred,SO2_y)),sqrt(Mae(O3_pred,O3_y))]
ll_pred = [PM_pred,PM10_pred,NO2_pred,SO2_pred,O3_pred]
ll_y = [PM_y,PM10_y,NO2_y,SO2_y,O3_y]
l = ["PM","PMT","NO2","SO2","O3"]
def plotting_all_data(l):
  for i in range(0,5):
    j = ll_y[i]
    k = ll_pred[i]
    plt.plot(j, color ='tab:blue',label='Actual '+l[i]) 
    plt.plot(k, color ='tab:orange', label="Predicted "+l[i])
    plt.title('Comparsion of Actual versus Predicted data in LSTM') 
    plt.legend(loc="upper right")
    plt.show()
plotting_all_data(l)
```

```python
data = {'MAE':mae,'RMSE':rmse}
mae_rmse_df = pd.DataFrame(data,index=l)
mae_rmse_df
```
